# Covid 19 Prediction

### Importing the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing the dataset of confirmed cases

```python
confirmed_cases = pd.read_csv("Confirmed_Cases.txt",delimiter=',')
confirmed_cases.head()
```

### Importing the dataset of deaths

```python
deaths = pd.read_csv("Deaths.txt",delimiter=',')
deaths.head()
```

### Importing the dataset of recoveries

```python
recovered = pd.read_csv("Recovered_Cases.txt",delimiter=',')
recovered.head()
```

### EDA

#### For Confirmed cases:-

```python
confirmed_cases = confirmed_cases.groupby(['Country/Region']).sum()
```

```python
confirmed_cases.drop(['Lat','Long'],axis=1,inplace=True)
```

```python
confirmed_cases.head()
```

#### For Deaths:-

```python
deaths = deaths.groupby(['Country/Region']).sum()
```

```python
deaths.drop(['Lat','Long'],axis=1,inplace=True)
```

```python
deaths.head()
```

#### For Recoveries:-

```python
recovered = recovered.groupby(['Country/Region']).sum()
```

```python
recovered.drop(['Lat','Long'],axis=1,inplace=True)
```

```python
recovered.head()
```

### For Confirmed Cases:

#### Splitting the data into training and test set

```python
import sklearn
from sklearn.model_selection import train_test_split


xc = confirmed_cases.drop('8/11/20',axis=1)
yc = confirmed_cases['8/11/20']

xc_train,xc_test,yc_train,yc_test = train_test_split(xc,yc,test_size=0.2,random_state=26)
```

#### Fitting the linear regression model

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(xc_train,yc_train)
```

#### Accuracy of the model

```python
lr.score(xc_test,yc_test)
```

#### Predictions

```python
yc_pred = lr.predict(xc_test)
```

```python
xc_test['Actual 8/11/20'] = yc_test
xc_test['Predicted 8/11/20'] = yc_pred
xc_test.head()
```

### For Deaths:

#### Splitting the dataset into training and test set

```python
xd = deaths.drop('8/11/20',axis=1)
yd = deaths['8/11/20']

xd_train,xd_test,yd_train,yd_test = train_test_split(xd,yd,test_size=0.2,random_state=26)
```

#### Fitting the linear Regression model

```python
lrd = LinearRegression()
lrd.fit(xd_train,yd_train)
```

#### Accuracy of the model

```python
lrd.score(xd_test,yd_test)
```

#### Prediction

```python
yd_pred = lrd.predict(xd_test)
```

```python
xd_test['Actual 8/11/20'] = yd_test
xd_test['Predicted 8/11/20'] = yd_pred
xd_test.head()
```

### For Recoveries

#### Splitting the dataset into training and test set

```python
xr = recovered.drop('8/11/20',axis=1)
yr = recovered['8/11/20']

xr_train,xr_test,yr_train,yr_test = train_test_split(xr,yr,test_size=0.2,random_state=26)
```

#### Fitting the Linear Regression model

```python
lrr = LinearRegression()
lrr.fit(xr_train,yr_train)
```

#### Accuracy of the model

```python
lrr.score(xr_test,yr_test)
```

#### Predictions

```python
yr_pred = lrr.predict(xr_test)
```

```python
xr_test['Actual 8/11/20'] = yr_test
xr_test['Predicted 8/11/20'] = yr_pred
xr_test.head()
```
