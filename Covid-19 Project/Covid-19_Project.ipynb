## For confirmed cases- 

```python
import sklearn
from sklearn.model_selection import train_test_split


xc = confirmed_cases.drop('8/11/20',axis=1)
yc = confirmed_cases['8/11/20']

xc_train,xc_test,yc_train,yc_test = train_test_split(xc,yc,test_size=0.2,random_state=26)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(xc_train,yc_train)
```

```python
lr.score(xc_test,yc_test)
```

```python
yc_pred = lr.predict(xc_test)
```

```python
xc_test['Actual 8/11/20'] = yc_test
xc_test['Predicted 8/11/20'] = yc_pred
xc_test.head()
```
